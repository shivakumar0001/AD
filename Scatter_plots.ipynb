```python
# prompt: import stastistics

import statistics
import numpy as np

stock_prices = [100,102,103,98,105, 101,97,46,108,56]
variance = statistics.variance(stock_prices)
print(f"variance of stock prices: {variance}")

numpy_variance = np.var(stock_prices)
print(f"numpy variance of stock prices: {numpy_variance}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Temperature': [25, 35, 56, 45, 65, 76, 34, 67, 55, 45, 35, 25],
    'Ice_Cream_Sales': [200, 250, 300, 350, 400, 420, 430, 450, 280, 310, 370, 560],
    'Advertising_Budget': [50, 56, 76, 87, 67, 56, 78, 98, 45, 67, 45, 34],
    'Customer_Count' : [80, 67, 52, 25, 56, 25, 58, 25, 85, 96, 65, 75]
}
df = pd.DataFrame(data)

sns.pairplot(df, diag_kind='kde', markers="o", plot_kws={'alpha': 0.45})
plt.suptitle("Pair Plot of Ice Cream Sales Data", y=1.02, fontsize=16)
plt.show()
```
