```python
import matplotlib.pyplot as plt
data = [5,7,7,8,9,2,4,6,7,8,9,10,11,11,12]
plt.hist(data, bins=10, edgecolor='white')
plt.title('Simple Histogram Example')
plt.xlabel('Numbers')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
datf = pd.DataFrame({"Season 1": [7,7,5,5,6,3],
                     "Season 2": [1,2,8,4,9,5]})
p = sns.histplot(data = datf)
p.set(xlabel="X label Value",ylabel="Y label Value")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(42)
data = np.random.randint(20,81,1000)
plt.hist(data, bins=15,edgecolor='white',color='red')
plt.title('Histogram of cancer Patients Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'semester': ['Sem1', 'Sem1', 'Sem1', 'Sem1', 'Sem1', 'Sem2', 'Sem2', 'Sem2', 'Sem2', 'Sem2'],
    'hours_studied': [10, 15, 12, 8, 14, 20, 14, 13, 16, 34]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 6))
sns.boxplot(x='semester', y='hours_studied', data=df)
plt.title('Box Plot of Hours Studied by Semester')
plt.xlabel('Semester')
plt.ylabel('Hours Studied')
plt.show()
```
